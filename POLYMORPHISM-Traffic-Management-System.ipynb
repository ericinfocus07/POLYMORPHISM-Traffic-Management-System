```python

class TrafficDevice:
    def activate(self):
        print("Traffic device activated.")



class TrafficLight(TrafficDevice):
    def activate(self):
        print("Traffic Light: Switching lights (Red → Yellow → Green).")


class SpeedCamera(TrafficDevice):
    def activate(self):
        print("Speed Camera: Monitoring vehicle speed.")


class PedestrianSignal(TrafficDevice):
    def activate(self):
        print("Pedestrian Signal: Walk signal is ON.")


class EmergencySiren(TrafficDevice):
    def activate(self):
        print("Emergency Siren: Giving priority to emergency vehicles.")



devices = [
    TrafficLight(),
    SpeedCamera(),
    PedestrianSignal(),
    EmergencySiren()
]

# Polymorphism
for device in devices:
    device.activate()
```
